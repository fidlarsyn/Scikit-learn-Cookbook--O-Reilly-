# **Introduction to SVMs**
Support Vector Machines (SVM) adalah model pembelajaran mesin yang sangat efektif untuk tugas klasifikasi dan regresi.
* **Hyperplane**: SVM bekerja dengan mencari hyperplane optimal yang berfungsi sebagai batas keputusan untuk memisahkan titik data ke dalam kelas-kelas yang berbeda.
* **Maksimalisasi Margin**: Tujuan utamanya adalah memaksimalkan jarak (margin) antara hyperplane dan titik data terdekat dari setiap kelas, yang dikenal sebagai support vectors. Margin yang lebih besar membantu model mencapai performa generalisasi yang lebih baik pada data baru.
* **Efektivitas**: SVM sangat efektif pada data dimensi tinggi atau ketika jumlah fitur lebih besar daripada jumlah sampel.

**Contoh Kode Dasar (SVM Classifier):**

```python
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris

# Memuat data
iris = load_iris()
X_train, X_test, y_train, y_test = train_test_split(
    iris.data, iris.target, test_size=0.3, random_state=2024)

# Inisialisasi model dengan kernel linear
svm_classifier = SVC(kernel='linear')

# Melatih model
svm_classifier.fit(X_train, y_train)

# Prediksi
y_pred = svm_classifier.predict(X_test)
```

# **Kernel functions and their applications**
Fungsi kernel memungkinkan SVM untuk menangani data yang tidak dapat dipisahkan secara linear dengan memetakan data asli ke ruang dimensi yang lebih tinggi (trik kernel).
* **Linear Kernel**: Digunakan jika data sudah dapat dipisahkan secara linear dalam ruang aslinya.
* **Polynomial Kernel**: Memetakan data menggunakan fungsi polinomial, berguna untuk fitur yang memiliki hubungan interaksi polinomial.
* **Radial Basis Function (RBF)**: Kernel yang paling umum digunakan karena kemampuannya menangani hubungan fitur yang sangat kompleks.

# **Tuning SVM parameters**
Performa SVM sangat bergantung pada pemilihan hyperparameter yang tepat melalui teknik seperti Grid Search.
* **Parameter C**: Mengontrol trade-off antara lebar margin dan kesalahan klasifikasi. Nilai C yang tinggi berarti model kurang toleran terhadap kesalahan (margin lebih sempit), sedangkan C rendah menciptakan margin yang lebih lebar namun mengizinkan beberapa misklasifikasi (soft margin).
* **Gamma**: Spesifik untuk kernel RBF; mendefinisikan seberapa jauh pengaruh satu contoh pelatihan mencapai.


**Contoh Kode Tuning (Grid Search):**

```python
from sklearn.model_selection import GridSearchCV

# Definisi parameter grid
param_grid = {
    'C': [0.1, 1, 10],
    'kernel': ['linear', 'rbf', 'poly'],
    'degree': [14-16] # Untuk kernel poly
}

# Pencarian parameter terbaik
grid_search = GridSearchCV(SVC(), param_grid, cv=5)
grid_search.fit(X_train, y_train)

print("Parameter Terbaik:", grid_search.best_params_)
```

# **SVMs in high-dimensional spaces**
SVM sangat efisien dalam menangani dataset lebar (wide datasets), di mana jumlah fitur jauh lebih banyak daripada jumlah sampel.
* **Kernel Trick**: Memungkinkan SVM beroperasi tanpa menghitung koordinat data secara eksplisit di ruang dimensi tinggi, yang menghemat biaya komputasi.
* **Regularisasi**: Dalam ruang dimensi tinggi, teknik regularisasi L1 atau L2 sangat krusial untuk mencegah overfitting agar model tidak mempelajari fitur yang tidak relevan.

# **Evaluating SVM models**
Evaluasi mendalam diperlukan untuk memahami seberapa baik model menggeneralisasi data baru.
* **Metrics**: Selain akurasi, metrik seperti Presisi, Recall, dan F1-score memberikan gambaran performa per kelas.
* **Kurva ROC & AUC**: Memvisualisasikan kemampuan model untuk membedakan antar kelas pada berbagai ambang batas keputusan. Skor AUC (Area Under Curve) yang mendekati 1 (atau 100%) menunjukkan kemampuan pemisahan kelas yang sangat baik.


**Contoh Kode Evaluasi:**

```python
from sklearn.metrics import classification_report, roc_curve, auc
import matplotlib.pyplot as plt

# Inisialisasi model dengan probabilitas aktif
svm_model = SVC(kernel='linear', probability=True)
svm_model.fit(X_train, y_train)

# Mendapatkan skor keputusan/probabilitas
y_prob = svm_model.predict_proba(X_test)[:, 1]

# Menghitung AUC
fpr, tpr, thresholds = roc_curve(y_test, y_prob, pos_label=1)
roc_auc = auc(fpr, tpr)
print(f"ROC AUC: {roc_auc:.2f}")

# Classification Report
print(classification_report(y_test, y_pred))
```
