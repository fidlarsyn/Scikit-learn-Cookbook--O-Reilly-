# **Introduction to Linear Models**
Model linear adalah fondasi teknik pemodelan prediktif yang mencari hubungan matematis antara variabel dependen (target) dan satu atau lebih variabel independen (fitur). Bentuk paling sederhana adalah Ordinary Least Squares (OLS) yang direpresentasikan sebagai garis lurus.

**Contoh Kode Dasar (Linear Regression):**

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

# Inisialisasi dan melatih model
linear_model = LinearRegression()
linear_model.fit(X_train, y_train)

# Membuat prediksi
y_pred = linear_model.predict(X_test)

# Evaluasi performa
print(f'MSE: {mean_squared_error(y_test, y_pred):.2f}')
print(f'R-squared: {r2_score(y_test, y_pred):.2f}')
```

# **Ridge and Lasso Regression**
Teknik ini meningkatkan performa model linear melalui regularisasi, yang menambahkan penalti pada fungsi kerugian (loss function) untuk mencegah overfitting.
* **Ridge Regression (L2):** Menambahkan penalti berdasarkan kuadrat besarnya koefisien. Ini menyusutkan koefisien mendekati nol tetapi tidak menghapusnya sama sekali.
* **Lasso Regression (L1):** Menambahkan penalti berdasarkan nilai absolut koefisien. Lasso dapat melakukan seleksi fitur otomatis dengan menyusutkan beberapa koefisien tepat menjadi nol.

**Contoh Kode Ridge & Lasso:**

```python
from sklearn.linear_model import Ridge, Lasso

# Ridge Regression (Alpha mengontrol kekuatan regulasi)
ridge_model = Ridge(alpha=1.0)
ridge_model.fit(X_train, y_train)

# Lasso Regression (Bermanfaat untuk dataset dengan fitur redundan)
lasso_model = Lasso(alpha=10.0, max_iter=10000)
lasso_model.fit(X_train, y_train)
```

# **ElasticNet and Regularization**
ElasticNet adalah hibrida yang menggabungkan kekuatan penalti L1 (Lasso) dan L2 (Ridge). Teknik ini sangat efektif ketika terdapat banyak fitur yang saling berkorelasi tinggi (multikolinearitas).

**Contoh Kode ElasticNet:**

```python
from sklearn.linear_model import ElasticNet

# l1_ratio mengontrol campuran antara Lasso (1) dan Ridge (0)
elastic_model = ElasticNet(alpha=1.0, l1_ratio=0.5, max_iter=10000)
elastic_model.fit(X_train, y_train)
```

# **Regularization Theory and Practice**
Regularisasi bertujuan untuk menyeimbangkan antara kompleksitas model dan seberapa baik model tersebut mencocokkan data (goodness of fit).
* **Mencegah Overfitting**: Tanpa regulasi, model mungkin menghafal kebisingan (noise) pada data latih.
* **Trade-off Bias-Variance**: Regularisasi sedikit meningkatkan bias (karena membatasi model) tetapi secara signifikan mengurangi varians (sensitivitas terhadap fluktuasi data latih), yang sering kali menghasilkan generalisasi yang lebih baik pada data baru.

* **Polynomial Regression**: Mentransformasi fitur menjadi istilah polinomial (misal: x^2, x^3) untuk menangkap tren kurva.
* **Spline Interpolation**: Membagi data menjadi segmen-segmen dan memasang polinomial pada setiap bagian, menciptakan kurva yang halus dan kontinu.
