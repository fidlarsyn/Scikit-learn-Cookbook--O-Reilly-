# **Overview of logistic regression**
Regresi logistik adalah metode statistik fundamental untuk masalah klasifikasi biner. Berbeda dengan regresi linear yang memprediksi nilai kontinu, regresi logistik memprediksi probabilitas bahwa titik data input termasuk dalam kategori tertentu menggunakan fungsi sigmoid (logistik) yang menghasilkan nilai antara 0 dan 1.

**Contoh Kode Dasar:**

```python
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split

# Memuat data
data = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(data.data, data.target, random_state=2024)

# Inisialisasi dan melatih model
model = LogisticRegression(max_iter=10000)
model.fit(X_train, y_train)

# Prediksi
y_pred = model.predict(X_test)
```

# **Multiclass classification techniques**
Regresi logistik dapat diperluas untuk menangani lebih dari dua kelas menggunakan dua strategi utama:
* **One-vs-Rest (OvR)**: Melatih satu pengklasifikasi biner terpisah untuk setiap kelas.
* **Multinomial (SoftMax)**: Memodelkan probabilitas setiap kelas secara langsung dan mengoptimalkan koefisien untuk semua kelas secara bersamaan.

**Contoh Kode Multikelas:**

```python
# Strategi OvR
ovr_model = LogisticRegression(multi_class='ovr', solver='liblinear')
ovr_model.fit(X_train, y_train)

# Strategi Multinomial (SoftMax)
multinomial_model = LogisticRegression(multi_class='multinomial', solver='lbfgs')
multinomial_model.fit(X_train, y_train)
```

# **Regularization in logistic regression**
Regularisasi digunakan untuk mencegah overfitting dengan menambahkan penalti pada koefisien yang terlalu besar.
* **Ridge (L2)**: Menyusutkan koefisien mendekati nol.
* **Lasso (L1)**: Dapat menghasilkan model yang jarang (sparse) dengan memaksa beberapa koefisien menjadi tepat nol, sehingga berfungsi sebagai seleksi fitur.
* **Parameter C**: Mengontrol kekuatan regularisasi (kebalikan dari kekuatan penalti); nilai C yang lebih kecil berarti regularisasi yang lebih kuat.

# **Multilabel classification concepts**
Dalam klasifikasi multilabel, setiap instans dapat memiliki beberapa label sekaligus (misalnya, satu artikel berita bisa berkategori "Politik" dan "Teknologi"). scikit-learn menangani ini biasanya dengan melatih model terpisah untuk setiap label atau menggunakan kelas pembungkus MultiOutputClassifier.

# **Model evaluation metrics**
Evaluasi yang tepat sangat penting untuk memahami kemampuan generalisasi model:
* **Presisi (Precision)**: Proporsi identifikasi positif yang benar-benar benar.
* **Recall**: Proporsi positif aktual yang diidentifikasi dengan benar.
* **F1-score**: Rata-rata harmonik antara presisi dan recall untuk keseimbangan metrik.
* **Kurva ROC & AUC**: Memvisualisasikan trade-off antara true positive rate dan false positive rate pada berbagai ambang batas keputusan.

# **Implementation in scikit-learn**
Berikut adalah contoh alur kerja evaluasi lengkap menggunakan metrik-metrik di atas:

```python
from sklearn.metrics import classification_report, roc_curve, auc
import matplotlib.pyplot as plt

# Mendapatkan probabilitas prediksi
y_prob = model.predict_proba(X_test)[:, 1]

# Menghitung metrik ROC
fpr, tpr, thresholds = roc_curve(y_test, y_prob)
roc_auc = auc(fpr, tpr)

# Menampilkan Laporan Klasifikasi
print(classification_report(y_test, y_pred))

# Plot Kurva ROC
plt.plot(fpr, tpr, label=f'ROC curve (area = {roc_auc:.2f})')
plt.plot([15], 'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend(loc="lower right")
plt.show()
```
