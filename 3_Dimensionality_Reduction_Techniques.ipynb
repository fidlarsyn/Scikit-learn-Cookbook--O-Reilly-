# **Introduction to Dimensionality Reduction**
Pengurangan dimensi adalah proses mengurangi jumlah fitur (dimensi) dalam dataset sambil tetap mempertahankan sebanyak mungkin informasi yang relevan. Teknik ini fundamental dalam membangun alur kerja ML yang kuat karena beberapa alasan utama:
* **Menyederhanakan Data**: Memudahkan interpretasi dan analisis data dimensi tinggi.
* **Mengurangi Biaya Komputasi**: Fitur yang lebih sedikit berarti waktu pelatihan model lebih cepat dan konsumsi sumber daya lebih rendah.
* **Meningkatkan Performa**: Membantu memitigasi risiko overfitting dengan menghilangkan fitur yang tidak relevan atau redundan.
* **Meningkatkan Visualisasi**: Memungkinkan proyeksi data dimensi tinggi ke dalam 2D atau 3D untuk inspeksi visual.

# **Transforming Datasets with PCA)**
Principal Component Analysis (PCA) adalah teknik tanpa pengawasan (unsupervised) yang menyederhanakan dataset dengan menangkap variansi terbesar dalam data melalui beberapa dimensi pertama yang disebut komponen utama.

Contoh Kode PCA:
**bold text**

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.pipeline import Pipeline
from sklearn.datasets import load_wine
import matplotlib.pyplot as plt

# Memuat data
wine = load_wine()
df_wine = wine.data

# Membangun pipeline: Standarisasi -> PCA
pca_pipeline = Pipeline([
    ('scaler', StandardScaler()),       # Wajib standarisasi sebelum PCA
    ('pca', PCA(n_components=2))        # Reduksi ke 2 dimensi
])

# Melatih dan mentransformasi data
X_pca = pca_pipeline.fit_transform(df_wine)

# Visualisasi variansi yang dijelaskan
pca_step = pca_pipeline.named_steps['pca']
print(f"Explained variance ratio: {pca_step.explained_variance_ratio_\}")
```

# **Maximizing Class Separability with LDA**
Berbeda dengan PCA, Linear Discriminant Analysis (LDA) adalah teknik terawasi (supervised). LDA bertujuan untuk memaksimalkan jarak antar kelas yang berbeda sambil meminimalkan variansi di dalam setiap kelas.

**Contoh Kode LDA:**

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.pipeline import Pipeline

# Membangun pipeline untuk LDA
lda_pipeline_wine = Pipeline([
    ('scaler', StandardScaler()),
    ('lda', LinearDiscriminantAnalysis(n_components=2))
])

# LDA membutuhkan label target (y_wine) karena bersifat supervised
X_lda_wine = lda_pipeline_wine.fit_transform(wine.data, wine.target)
```

# **t-SNE and Data Visualization**
t-Distributed Stochastic Neighbor Embedding (t-SNE) adalah teknik non-linear yang sangat efektif untuk memetakan data dimensi tinggi ke ruang 2D atau 3D. Teknik ini unggul dalam mengungkap struktur lokal dan pengelompokan (klaster) yang kompleks.

**Contoh Kode t-SNE (Dataset Digits):**

```python
from sklearn.manifold import TSNE
from sklearn.datasets import load_digits

digits = load_digits()

# Membangun alur kerja t-SNE
tsne_pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('tsne', TSNE(n_components=2, random_state=2024))
])

# t-SNE mentransformasi data mentah gambar digit
X_tsne = tsne_pipeline.fit_transform(digits.data)
```

# **Selecting the Right Technique**
Buku ini menyediakan panduan untuk memilih teknik berdasarkan tujuan dan jenis data:
* **LDA**: Pilih jika Anda memiliki data berlabel dan ingin memaksimalkan pemisahan antar kelas untuk klasifikasi.
* **PCA**: Pilih untuk prapemrosesan umum, pengurangan noise, atau ekstraksi fitur pada data tanpa label.PCA bersifat linier dan cepat.
* **t-SNE**: Gunakan khusus untuk visualisasi eksploratif guna mengungkap pola non-linear dan struktur klaster lokal. t-SNE tidak disarankan untuk prapemrosesan data baru karena komputasinya yang intensif.

# **Impact on Model Performance**
Pengurangan dimensi melibatkan trade-off (pertukaran) yang harus dipahami oleh praktisi ML:
* **Kompleksitas vs Interpretabilitas**: Fitur yang lebih sedikit menghasilkan model yang lebih sederhana dan mudah dijelaskan, namun pengurangan dimensi yang berlebihan dapat membuang informasi kritis.
* **Waktu Pelatihan vs Akurasi**: Dataset yang lebih kecil mempercepat pelatihan secara signifikan, namun ada potensi penurunan akurasi jika fitur penting hilang.
* **Overfitting vs Generalisasi**: Teknik ini membantu mencegah model mempelajari "noise" (derau), yang sering kali meningkatkan kemampuan model untuk bekerja pada data baru yang belum pernah dilihat (generalization). Namun, jika terlalu banyak dimensi dibuang, model bisa menjadi terlalu sederhana (underfitting).
