# **Prinsip Desain scikit-learn**
Desain scikit-learn didasarkan pada empat prinsip utama:
* Konsistensi: Semua objek berbagi antarmuka yang seragam (misal: penggunaan fit() untuk melatih semua jenis model).
* Kesederhanaan: Fokus pada algoritme machine learning tanpa menambahkan kompleksitas yang tidak perlu.
* Modularitas: Komponen seperti pemrosesan data, model, dan pipeline dapat dikombinasikan dengan fleksibel.
* Reusabilitas: Memungkinkan pembuatan komponen kustom yang tetap kompatibel dengan ekosistem scikit-learn.

# **Memahami Estimator (Model)**
Estimator adalah objek yang mengimplementasikan algoritme pembelajaran. Dua metode utamanya adalah:
* fit(): Digunakan untuk melatih model dengan mempelajari pola dari data.
* predict(): Digunakan untuk membuat prediksi pada data baru berdasarkan model yang telah dilatih.

**Contoh Kode (Linear Regression):**

```python
from sklearn.linear_model import LinearRegression
import numpy as np

# Data fitur (X) dan target (y)
X = np.array([[6], [7], [8], [9], [10]])
y = np.array([1, 2, 3, 3.5, 5])

# Inisialisasi dan melatih model
model = LinearRegression()
model.fit(X, y) # Menggunakan fit()

# Membuat prediksi
X_new = np.array([[1], [11]])
predictions = model.predict(X_new)
print("Prediksi:", predictions)
```

# **Transformers dan Metode `transform()`**
Transformers adalah objek yang digunakan untuk membersihkan atau memanipulasi data sebelum dimasukkan ke model [13].
*   Metode `fit()` menghitung parameter (misal: rata-rata untuk penskalaan) [13].
Metode `transform()` menerapkan manipulasi tersebut ke data

**Contoh Kode (StandardScaler):**

```python
from sklearn.preprocessing import StandardScaler

X = np.array([[6, 7], [8, 9], [1, 10]])
scaler = StandardScaler()

# Menghitung parameter dan mengubah data
X_scaled = scaler.fit_transform(X)
```

# **Atribut dan Metode Umum**
Setelah pelatihan (`fit()`), scikit-learn menyimpan hasil pembelajaran dalam atribut yang diakhiri dengan garis bawah (`_`).
*   `coef_`: Menyimpan koefisien bobot model.
*   `intercept_`: Menyimpan titik potong bias.
*   `score()`: Metode cepat untuk mengevaluasi performa (mengembalikan R² untuk regresi atau akurasi untuk klasifikasi).

**Contoh Kode:**

```python
print("Koefisien:", model.coef_)
print("Intercept:", model.intercept_)
print("Skor Model (R²):", model.score(X, y))
```

# **Hyperparameter Tuning**
Hyperparameter memberi tahu model *bagaimana* cara belajar. scikit-learn menyediakan metode seragam untuk mengelola parameter ini:
*   `get_params()`: Melihat parameter model saat ini.
*   `set_params()`: Mengubah parameter sebelum proses pelatihan dimulai.

**Contoh Program:**

```python
from sklearn.ensemble import RandomForestClassifier

# 1. Inisialisasi model
model = RandomForestClassifier()

# 2. Mengatur hyperparameter sebelum pelatihan menggunakan set_params()
model.set_params(n_estimators=100, max_depth=10, random_state=42)

# 3. Mengecek parameter yang telah diperbarui
print(model.get_params())
```

# **Pipeline dan Metadata Routing**
* Pipelines: Memungkinkan otomatisasi alur kerja dengan merantai beberapa langkah (seperti penskalaan dan pemodelan) ke dalam satu objek tunggal agar prosesnya dapat diulang dengan mudah.
* Metadata Routing: Fitur canggih yang memungkinkan pengguna mengontrol bagaimana informasi tambahan (seperti bobot sampel atau label grup) diteruskan di antara langkah-langkah dalam pipeline.


**Contoh Program:**

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression

# Membangun pipeline: (Nama langkah, Objek)
pipeline = Pipeline([
    ("imputer", SimpleImputer(strategy="mean")), # Langkah 1: Isi data hilang
    ("scaler", StandardScaler()),                # Langkah 2: Skalakan fitur
    ("clf", LogisticRegression())                # Langkah 3: Model klasifikasi
])

# Melatih seluruh alur kerja sekaligus
pipeline.fit(X_train, y_train)

# Membuat prediksi yang otomatis melewati proses imputer dan scaler
predictions = pipeline.predict(X_test)
```

Teknik ini membagi data menjadi beberapa bagian (folds) untuk mendapatkan estimasi performa yang lebih stabil.

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()

# Melakukan 5-fold cross-validation
# Mengembalikan skor akurasi untuk setiap iterasi
scores = cross_val_score(model, X, y, cv=5, scoring='accuracy')

print(f"Rata-rata akurasi: {scores.mean():.3f}")
```
