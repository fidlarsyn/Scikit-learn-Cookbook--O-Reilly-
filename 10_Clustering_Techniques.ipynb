# **Introduction to clustering**
Klastering adalah proses membagi kumpulan data menjadi beberapa kelompok (klaster) sehingga titik data dalam satu kelompok memiliki kemiripan yang tinggi satu sama lain dibandingkan dengan titik data di kelompok lain.
* **Prinsip Kerja**: Algoritme menghitung kemiripan berdasarkan metrik jarak (seperti Euclidean) atau ukuran kepadatan.
* **Tipe Utama**: Centroid-based: Membentuk klaster di sekitar titik pusat (contoh: K-means).
* **Connectivity-based**: Membangun klaster menggunakan keterkaitan hierarkis (contoh: Agglomerative Clustering).
* **Density-based**: Mendeteksi klaster berdasarkan kepadatan wilayah dan mampu mengidentifikasi noise (contoh: DBSCAN).

# **K-means Clustering**
K-means adalah algoritme klastering paling populer yang membagi data ke dalam k klaster. Algoritme ini bekerja secara iteratif: menginisialisasi centroid, menetapkan titik ke centroid terdekat, dan memperbarui posisi centroid berdasarkan rata-rata titik di dalamnya.

**Contoh Kode K-means & Metode Elbow:**

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# 1. Inisialisasi model dengan 4 klaster
kmeans = KMeans(n_clusters=4, random_state=2024)
y_kmeans = kmeans.fit_predict(X) # X adalah data yang sudah diskalakan

# 2. Mencari jumlah klaster optimal dengan Metode Elbow (Inertia)
inertia = []
for k in range(1, 10):
    km = KMeans(n_clusters=k, random_state=2024)
    km.fit(X)
    inertia.append(km.inertia_)

plt.plot(range(1, 10), inertia, 'bo-')
plt.title("Elbow Method for Optimal k")
plt.show()
```

# **Hierarchical Clustering**
Metode ini membangun struktur klaster bertingkat. Salah satu bentuknya adalah Agglomerative Clustering, yang memulai dengan setiap titik sebagai klaster sendiri dan menggabungkannya secara bertahap. Hasilnya dapat divisualisasikan menggunakan Dendrogram.

**Contoh Kode Klastering Hierarkis:**

```python
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage

# 1. Implementasi model
agg = AgglomerativeClustering(n_clusters=4)
y_agg = agg.fit_predict(X)

# 2. Membuat Dendrogram untuk melihat struktur hierarkis
linked = linkage(X, method='ward')
dendrogram(linked)
plt.title("Dendrogram Hierarchical Clustering")
plt.show()
```

# **Density-based clustering with DBSCAN**
DBSCAN (Density-Based Spatial Clustering of Applications with Noise) sangat efektif untuk data dengan bentuk yang tidak beraturan (non-convex) dan mampu mendeteksi pencilan (outliers) sebagai noise.
* **Parameter Utama**: eps (radius lingkungan) dan min_samples (jumlah minimum titik untuk membentuk wilayah padat).

**Contoh Kode DBSCAN:**

```python
from sklearn.cluster import DBSCAN
from sklearn.neighbors import NearestNeighbors
import numpy as np

# 1. Menentukan eps optimal menggunakan k-distance plot
neighbors = NearestNeighbors(n_neighbors=5).fit(X)
distances, indices = neighbors.kneighbors(X)
distances = np.sort(distances[:, 4])
plt.plot(distances) # Cari titik 'siku' (elbow) pada plot

# 2. Implementasi DBSCAN
dbscan = DBSCAN(eps=0.3, min_samples=5)
y_dbscan = dbscan.fit_predict(X)
```

# **Cluster Evaluation Metrics**
Karena data tidak memiliki label, evaluasi dilakukan untuk mengukur seberapa baik klaster terbentuk:
* **Silhouette Score**: Mengukur seberapa dekat sebuah titik dengan klasternya sendiri dibandingkan dengan klaster lain (Skor 1 sangat baik, -1 buruk).
* **Davies-Bouldin Index**: Menilai pemisahan dan kepadatan klaster (Skor lebih rendah lebih baik).
* **Adjusted Rand Index (ARI)**: Digunakan jika label asli (ground truth) tersedia untuk validas.

# **Cluster evaluation metrics**
Pemilihan algoritme sangat bergantung pada struktur data dan tingkat noise. Berikut panduan pemilihannya:
* Klaster cembung (bulat/convex) -> K-means
* Klaster non-cembung (berbentuk unik/crescent) -> DBSCAN atau Spectral Clustering
* Adanya banyak noise atau pencilan -> DBSCAN
* Dataset kecil dengan struktur hierarkis -> Hierarchical Clustering
* Dataset sangat besar -> K-means atau MiniBatchKMeans


# **Advanced Techniques**
* Pustaka scikit-learn juga menawarkan teknik yang lebih fleksibel untuk tugas-tugas yang bernuansa:
* **Spectral Clustering**: Efektif untuk klaster non-cembung dengan cara memetakan data ke ruang fitur berdimensi lebih rendah sebelum menerapkan K-means.
* **Gaussian Mixture Models (GMM)**: Metode klastering "lunak" (soft clustering) yang mengasumsikan data dihasilkan dari campuran beberapa distribusi Gaussian; setiap titik diberikan probabilitas milik setiap klaster.
