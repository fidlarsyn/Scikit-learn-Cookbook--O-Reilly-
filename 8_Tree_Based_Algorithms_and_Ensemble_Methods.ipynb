# **Introduction to decision trees**
Pohon keputusan bekerja dengan memecah data secara rekursif berdasarkan nilai fitur untuk menciptakan struktur seperti pohon yang terdiri dari node dan cabang.
* **Interpretabilitas**: Sangat mudah dipahami karena alur keputusannya dapat divisualisasikan.
* **Kriteria Split**: Secara default, scikit-learn menggunakan Gini impurity untuk menentukan pemisahan data yang paling homogen di setiap tingkat.


**Contoh Kode Dasar:**

```python
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt

# Memuat data
iris = load_iris()
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, random_state=2024)

# Inisialisasi dan melatih model
clf = DecisionTreeClassifier(random_state=2024)
clf.fit(X_train, y_train)

# Visualisasi pohon
plt.figure(figsize=(15,10))
plot_tree(clf, feature_names=iris.feature_names, class_names=iris.target_names, filled=True)
plt.show()
```

# **Random forests and bagging**
Random Forests adalah metode ensemble yang melatih banyak pohon keputusan secara paralel pada subset data yang berbeda menggunakan teknik bagging (bootstrap aggregating).
* **Keunggulan**: Mengurangi risiko overfitting dan meningkatkan stabilitas model dibandingkan satu pohon tunggal.
* **Voting**: Prediksi akhir ditentukan berdasarkan suara mayoritas dari seluruh pohon.

**Contoh Kode Random Forest:**

```python
from sklearn.ensemble import RandomForestClassifier

# Melatih Random Forest dengan 100 pohon (estimators)
rf_clf = RandomForestClassifier(n_estimators=100, random_state=2024)
rf_clf.fit(X_train, y_train)

# Melihat fitur yang paling berpengaruh
importances = rf_clf.feature_importances_
print("Fitur Penting:", importances)
```

# **Gradient Boosting Machines (GBM)**
Berbeda dengan Random Forest yang bekerja paralel, GBM membangun pohon secara sekuensial. Setiap pohon baru bertujuan untuk meminimalkan kesalahan (residual) dari model sebelumnya.
* **Boosting**: Teknik ini mencoba "mendorong" performa prediksi di setiap iterasi berturut-turut.
* **Karakteristik**: Sangat efektif untuk meningkatkan akurasi, namun lebih rentan terhadap overfitting jika parameter tidak disetel dengan hati-hati.

**Contoh Kode GBM:**

```python
from sklearn.ensemble import GradientBoostingClassifier

# Inisialisasi GBM
gbm_clf = GradientBoostingClassifier(n_estimators=100, learning_rate=0.1, random_state=2024)
gbm_clf.fit(X_train, y_train)
```

# **Hyperparameter tuning for trees and ensembles**
Penyetelan sangat penting untuk mencegah model menjadi terlalu kompleks atau terlalu sederhana.
* **max_depth**: Mengontrol kedalaman pohon.
* **min_samples_split**: Jumlah minimum sampel yang diperlukan untuk membagi sebuah node.
* **learning_rate**: (Untuk GBM) Menentukan seberapa besar kontribusi setiap pohon baru.


**Contoh Kode Tuning (Grid Search):**

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    'n_estimators': [13-15],
    'max_depth': [16-18],
    'learning_rate': [0.01, 0.1, 0.2]
}

grid_search = GridSearchCV(GradientBoostingClassifier(random_state=2024), param_grid, cv=5)
grid_search.fit(X_train, y_train)
print("Parameter Terbaik:", grid_search.best_params_)
```

# **Comparing ensemble methods**
Setiap metode memiliki karakteristik unik yang dapat disesuaikan dengan jenis masalahnya:
* **Bagging (Random Forest)**: Fokus pada pengurangan varians (model lebih stabil).
* **Boosting (GBM)**: Fokus pada pengurangan bias (model lebih akurat).
* **Stacking**: Menggabungkan berbagai algoritme (misal RF dan GBM) menggunakan meta-model untuk meningkatkan performa lebih jauh lagi.

**Contoh Kode Stacking:**

```python
from sklearn.ensemble import StackingClassifier
from sklearn.linear_model import LogisticRegression

# Menggabungkan RF dan GBM dengan Logistic Regression sebagai penggabung akhir
stacking_clf = StackingClassifier(
    estimators=[('rf', rf_clf), ('gb', gbm_clf)],
    final_estimator=LogisticRegression(),
    cv=5
)
stacking_clf.fit(X_train, y_train)
```
