## **Introduction to Distance Metrics**
Metrik jarak adalah komponen fundamental dalam berbagai algoritme machine learning yang mengandalkan konsep kemiripan (similarity) atau ketidakmiripan (dissimilarity) antar titik data. Memahami cara mengukur jarak antar titik dalam ruang fitur sangat penting untuk tugas-tugas seperti pengelompokan (clustering), klasifikasi, dan regresi. Dalam klasifikasi, kita sering membuat asumsi bahwa jika sebuah objek terlihat, berjalan, dan bersuara seperti bebek, maka kemungkinan besar itu adalah bebek


# **Understanding KNNs**
Algoritme k-Nearest Neighbors (KNN) adalah teknik pembelajaran terawasi (supervised learning) dasar yang digunakan untuk klasifikasi dan regresi.
* **Prinsip Kerja**: KNN mengestimasi karakteristik titik data baru berdasarkan titik-titik data lama yang paling dekat dengannya menggunakan metrik jarak
* **Pemungutan Suara (Voting)**: Untuk klasifikasi, KNN menetapkan label berdasarkan suara mayoritas dari k tetangga terdekatnya.
* **Parameter k**: Nilai k yang kecil membuat model sensitif terhadap derau (noise), sementara k yang besar menghaluskan prediksi namun mungkin mengabaikan pola lokal.


# **Distance Metrics Overview**
Ada tiga metrik jarak utama yang sering digunakan:
* **Euclidean Distance**: Jarak garis lurus standar dalam ruang n-dimensi (seperti burung terbang). Metrik ini paling umum digunakan tetapi sensitif terhadap penskalaan fitur.
* **Manhattan Distance (L1)**: Juga dikenal sebagai jarak blok kota, metrik ini mengukur jumlah perbedaan absolut antar koordinat (seperti menavigasi kisi jalan di kota). Metrik ini lebih kuat (robust) terhadap pencilan (outliers).
* **Minkowski Distance**: Generalisasi dari jarak Euclidean dan Manhattan dengan parameter p. Jika p=1, ini menjadi Manhattan; jika p=2, ini menjadi Euclidean.

# **Model Implementation**
scikit-learn menyediakan kelas KNeighborsClassifier yang sangat mudah diimplementasikan.

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris
from sklearn.metrics import accuracy_score

# Memuat dan membagi data
iris = load_iris()
X_train, X_test, y_train, y_test = train_test_split(
    iris.data, iris.target, test_size=0.2, random_state=2024)

# Inisialisasi model dengan 3 tetangga
knn = KNeighborsClassifier(n_neighbors=3)

# Melatih model (fit)
knn.fit(X_train, y_train)

# Membuat prediksi
y_pred = knn.predict(X_test)
print(f"Akurasi Awal: {accuracy_score(y_test, y_pred):.3f}")
```

# **Hyperparameter tuning in KNN**
Hyperparameter memberi tahu model bagaimana cara belajar. Teknik Grid Search digunakan untuk menguji berbagai kombinasi hyperparameter secara menyeluruh.

```python
from sklearn.model_selection import GridSearchCV

# Definisi grid hyperparameter
param_grid = {
    'n_neighbors': [12-16],
    'weights': ['uniform', 'distance'],
    'metric': ['euclidean', 'manhattan']
}

# Pencarian parameter terbaik dengan 5-fold cross-validation
grid_search = GridSearchCV(KNeighborsClassifier(), param_grid, cv=5, scoring='accuracy')
grid_search.fit(X_train, y_train)

print("Parameter Terbaik:", grid_search.best_params_)
print("Skor Cross-Validation Terbaik:", grid_search.best_score_)
```

# **Evaluating KNN Performance**
Evaluasi performa sangat penting untuk memahami seberapa baik model melakukan generalisasi pada data baru.
* **Kurva Pembelajaran (Learning Curves)**: Memplot performa terhadap ukuran dataset untuk mengidentifikasi underfitting atau overfitting.
* **Matriks Kebingungan (Confusion Matrix)**: Memberikan ringkasan visual jumlah prediksi benar dan salah untuk setiap kelas.
* **Classification Report**: Menyajikan metrik mendalam seperti Presisi (akurasi prediksi positif), Recall (sensitivitas dalam menemukan instans relevan), dan F1-score (keseimbangan antara presisi dan recall).


```python
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Evaluasi menggunakan model terbaik
best_knn = grid_search.best_estimator_
y_pred_best = best_knn.predict(X_test)

# Menampilkan Confusion Matrix
cm = confusion_matrix(y_test, y_pred_best)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=iris.target_names, yticklabels=iris.target_names)
plt.title('Confusion Matrix')
plt.show()

# Menampilkan Laporan Klasifikasi
print(classification_report(y_test, y_pred_best, target_names=iris.target_
```
