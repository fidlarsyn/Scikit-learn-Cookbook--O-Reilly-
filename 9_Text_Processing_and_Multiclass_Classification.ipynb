# **Introduction to text processing**
Sebagian besar data di dunia bersifat tidak terstruktur (seperti email dan dokumen), sehingga langkah pertama yang krusial adalah pembersihan dan konversi teks menjadi data numerik terstruktur.
* **Tokenisasi**: Memecah teks menjadi unit-unit kecil seperti kata atau kalimat (token).
* **Stopwords**: Kata-kata umum (seperti "the", "a", "an") yang sering dihapus karena memberikan sedikit nilai informasi untuk tugas klasifikasi.

**Contoh Kode Dasar (CountVectorizer):**

```python
from sklearn.feature_extraction.text import CountVectorizer

# Teks contoh
texts = ["Northern liberals are the chief supporters", "integration is key"]

# Inisialisasi dengan penghapusan stop words bahasa Inggris
vectorizer = CountVectorizer(stop_words='english')

# Transformasi teks menjadi matriks frekuensi token
X_vect = vectorizer.fit_transform(texts)
print(vectorizer.get_feature_names_out())
```

# **Text vectorization techniques**
Ada dua pendekatan utama yang dibahas untuk merepresentasikan teks secara numerik:
* **Bag of Words (BoW)**: Menghitung frekuensi kemunculan kata tanpa mempedulikan urutan atau tata bahasa.
* **TF-IDF (Term Frequency-Inverse Document Frequency)**: Memberikan bobot lebih tinggi pada kata-kata yang unik atau khas dalam sebuah dokumen dan bobot lebih rendah pada kata yang muncul terlalu sering di banyak dokumen.

**Contoh Kode Vektorisasi:**

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# TF-IDF Vectorization
tfidf_vectorizer = TfidfVectorizer(stop_words='english')
X_tfidf = tfidf_vectorizer.fit_transform(texts)
```

# **Feature extraction from text**
Selain frekuensi kata tunggal, fitur yang lebih kaya dapat diekstraksi untuk meningkatkan performa model:
* **N-gram**: Menangkap urutan kata (misal: bigram untuk menangkap konteks dua kata berurutan).
* **POS Tagging**: Mengkategorikan kata berdasarkan jenis tata bahasanya (kata benda, kata kerja, dll.) untuk memahami struktur sintaksis.

**Contoh Kode N-grams:**

```python
# Mengekstraksi unigram dan bigram (ngram_range 1 hingga 2)
ngram_vectorizer = CountVectorizer(ngram_range=(1, 2))
X_ngram = ngram_vectorizer.fit_transform(texts)
```

# **Multiclass classification strategies**
Ketika menghadapi masalah dengan lebih dari dua label kategori, scikit-learn menggunakan dua strategi utama:
* **One-vs-Rest (OvR)**: Melatih satu pengklasifikasi biner untuk setiap kelas terhadap semua kelas lainnya.
* **One-vs-One (OvO)**: Membangun pengklasifikasi untuk setiap pasangan kelas yang unik (lebih intensif secara komputasi namun seringkali lebih akurat pada dataset yang seimbang).
